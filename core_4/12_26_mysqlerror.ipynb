```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://health_user:strong_password_here@localhost:3306/HEALTH"
)
```

```python
import pandas as pd
from sqlalchemy import text

df = pd.read_sql("SELECT * FROM health_state_log", engine)

df_clean = (
    df.sort_values("id")
      .drop_duplicates(subset=["asset_id", "t_index", "source"], keep="first")
)

with engine.begin() as conn:
    conn.execute(text("TRUNCATE TABLE health_state_log"))

df_clean.to_sql(
    "health_state_log",
    con=engine,
    if_exists="append",
    index=False
)

print("health_state_log cleaned:", len(df), "→", len(df_clean))
```
