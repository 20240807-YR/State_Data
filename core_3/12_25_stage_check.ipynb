 Stage A — NASA → nasa_core.csv

```python
import pandas as pd

# NASA metadata
nasa_df = pd.read_csv("../data_csv/NASAmetadata.csv")

# discharge만
nasa_df = nasa_df[nasa_df["type"] == "discharge"].copy()

# Capacity 있는 행만
nasa_df = nasa_df.dropna(subset=["Capacity"])

# 공통 포맷
nasa_core = (
    nasa_df
    .assign(
        asset_id=lambda x: x["battery_id"].astype(str),
        t_index=lambda x: x.groupby("battery_id").cumcount(),
        state_value=lambda x: x["Capacity"]
    )
    [["asset_id", "t_index", "state_value"]]
)

# 저장
nasa_core.to_csv("../data_csv/nasa_core.csv", index=False)

print("Saved: nasa_core.csv")
```

Stage B — liBattery → libattery_core.csv

```python
import pandas as pd

li_df = pd.read_csv("../data_csv/liBattery_Data_Cleaned.csv")

li_core = (
    li_df
    .sort_values(["battery_id", "uid"])
    .assign(
        asset_id=lambda x: x["battery_id"].astype(str),
        t_index=lambda x: x.groupby("battery_id").cumcount(),
        state_value=lambda x: x["Capacity"]
    )
    [["asset_id", "t_index", "state_value"]]
)

li_core.to_csv("../data_csv/libattery_core.csv", index=False)

print("Saved: libattery_core.csv")
```

Stage C-1 — Synthetic Degradation → synthetic_degradation_core.csv

```python
import pandas as pd

synth_df = pd.read_csv("../data_csv/battery_degradation.csv")

synth_core = (
    synth_df
    .sort_values(["battery_id", "time"])
    .assign(
        asset_id=lambda x: x["battery_id"].astype(str),
        t_index=lambda x: x.groupby("battery_id").cumcount(),
        state_value=lambda x: x["rul"]
    )
    [["asset_id", "t_index", "state_value"]]
)

synth_core.to_csv("../data_csv/synthetic_degradation_core.csv", index=False)

print("Saved: synthetic_degradation_core.csv")
```

Stage C-2 — EV Synthetic → ev_synth_core.csv

```python
import pandas as pd

ev_df = pd.read_csv("../data_csv/ev_battery_synth.csv")

ev_core = (
    ev_df
    .sort_values(["battery_id", "charge_cycles"])
    .assign(
        asset_id=lambda x: x["battery_id"].astype(str),
        t_index=lambda x: x["charge_cycles"],
        state_value=lambda x: x["capacity_kWh"]
    )
    [["asset_id", "t_index", "state_value"]]
)

ev_core.to_csv("../data_csv/ev_synth_core.csv", index=False)

print("Saved: ev_synth_core.csv")
```
