```python
import pandas as pd

def inspect_csv(path, n=5):
    print("=" * 80)
    print(f"FILE: {path}")
    print("=" * 80)
    
    df = pd.read_csv(path)
    
    print("\n[Columns]")
    print(df.columns.tolist())
    
    print("\n[Head]")
    display(df.head(n))
    
    print("\n[Null count]")
    print(df.isna().sum())
    
    print("\n[Row count]")
    print(len(df))
    
    return df
```
