Case A 결과 logging (Prediction-based)

```python
with mlflow.start_run(run_name="CaseA_prediction_based"):
    mlflow.log_param("decision_type", "prediction_only")
    mlflow.log_param("input", "HDR + risk_group")

    stab_rate = df[df["intervention_flag"] == 1]["stabilized"].mean()
    false_rate = (
        (df["intervention_flag"] == 1) & (df["stabilized"] == False)
    ).mean()

    mlflow.log_metric("stabilization_rate", stab_rate)
    mlflow.log_metric("false_intervention_rate", false_rate)
```

Case B 결과 logging (μHSM-based)

```python
with mlflow.start_run(run_name="CaseB_muHSM_based"):
    mlflow.log_param("decision_type", "state_based")
    mlflow.log_param(
        "input",
        "HSI + HDR + RM + OBS"
    )

    stab_rate = df[df["intervention_flag"] == 1]["stabilized"].mean()
    false_rate = (
        (df["intervention_flag"] == 1) & (df["stabilized"] == False)
    ).mean()

    mlflow.log_metric("stabilization_rate", stab_rate)
    mlflow.log_metric("false_intervention_rate", false_rate)
```
